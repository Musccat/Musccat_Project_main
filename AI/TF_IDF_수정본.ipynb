# **패션디자인 1학년 4.0 소득분위 5**

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 학생 정보
student_info = {
    "학년": "1학년",
    "전공": "패션디자인",
    "성적": 4.0,
    "소득분위": 3,
    "기타": ["예술계열/예술대학"] # 지역, 계열 등 강조하고 싶은 점 작성
}

def check_criteria(criteria, student_info):
    score = 0

    # 학년 조건 비교 (~학년 이상도 포함)
    if re.search(r'(\d+학년)', criteria):
        required_year = re.findall(r'(\d+학년)', criteria)
        if any(int(year[0]) <= int(student_info["학년"][0]) for year in required_year):
            score += 2

    # 전공 조건 비교
    if re.search(student_info["전공"], criteria):
        score += 2.5

    # 성적 조건 비교
    if re.search(r'(성적|학점|평점) ([\d\.]+)', criteria):
        min_gpa = float(re.findall(r'(성적|학점|평점) ([\d\.]+)', criteria)[0][1])
        if student_info["성적"] >= min_gpa:
            score += 1

    # 소득분위 조건 비교 (장학금 기준의 소득분위보다 낮은지 체크)
    if re.search(r'소득분위 (\d+)', criteria):
        max_income_bracket = int(re.findall(r'소득분위 (\d+)', criteria)[0])
        if student_info["소득분위"] <= max_income_bracket:
            score += 2

    # 성적 우수의 기준 제시하여 비교 (4.0 이상이면 우수하다고 판단)
    if ("성적 우수" in criteria or "성적이 우수한" in criteria) and student_info["성적"] >= 4.0:
        score += 1

    # 가계곤란한 기준 제시하여 비교 (소득분위 4 이하인 경우 해당한다고 판단)
    if ("가계 곤란한" in criteria or "가정형편이 어려운" in criteria or "가정형편이 어렵고" in criteria) and student_info["소득분위"] <= 4:
        score += 1

    return score

# 사용자 프로필을 텍스트로 변환하는 함수
def student_info_to_text(info):
    return f"{info['학년']} {info['전공']} 성적 {info['성적']} 소득분위 {info['소득분위']} {' '.join(info['기타'])}"

# 학생 정보를 텍스트로 변환
student_text = student_info_to_text(student_info)

# 장학금 조건 텍스트로 변환
scholarship_texts = [scholarship["criteria"] for scholarship in scholarships]

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([student_text] + scholarship_texts)

# 코사인 유사도 계산
cosine_similarities = cosine_similarity(vectors[0:1], vectors[1:]).flatten()

# 유사도와 추가 점수를 합산하여 최종 점수 계산
final_scores = []
for i, scholarship in enumerate(scholarships):
    criteria_score = check_criteria(scholarship['criteria'], student_info)
    final_score = cosine_similarities[i]*50 + criteria_score
    final_scores.append((scholarship['name'], final_score, scholarship['criteria']))

# 최종 점수 순으로 정렬하여 추천
final_scores = sorted(final_scores, key=lambda x: x[1], reverse=True)

# 추천 장학금 목록 출력
print("추천 장학금 목록:")
for scholarship, score, criteria in final_scores[:4]:
    print(f"{scholarship} (점수: {score:.2f}) - 기준: {criteria}")
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 학생 정보
student_info = {
    "학년": "3학년",
    "전공": "컴퓨터공학",
    "성적": 4.2,
    "소득분위": 2,
    "기타": ["이공계"] # 지역, 계열 등 강조하고 싶은 점 작성
}

def check_criteria(criteria, student_info):
    score = 0

    # 학년 조건 비교 (~학년 이상도 포함)
    if re.search(r'(\d+학년)', criteria):
        required_year = re.findall(r'(\d+학년)', criteria)
        if any(int(year[0]) <= int(student_info["학년"][0]) for year in required_year):
            score += 2

    # 전공 조건 비교
    if re.search(student_info["전공"], criteria):
        score += 2.5

    # 성적 조건 비교
    if re.search(r'(성적|학점|평점) ([\d\.]+)', criteria):
        min_gpa = float(re.findall(r'(성적|학점|평점) ([\d\.]+)', criteria)[0][1])
        if student_info["성적"] >= min_gpa:
            score += 1

    # 소득분위 조건 비교 (장학금 기준의 소득분위보다 낮은지 체크)
    if re.search(r'소득분위 (\d+)', criteria):
        max_income_bracket = int(re.findall(r'소득분위 (\d+)', criteria)[0])
        if student_info["소득분위"] <= max_income_bracket:
            score += 2

    # 성적 우수의 기준 제시하여 비교 (4.0 이상이면 우수하다고 판단)
    if ("성적 우수" in criteria or "성적이 우수한" in criteria) and student_info["성적"] >= 4.0:
        score += 1

    # 가계곤란한 기준 제시하여 비교 (소득분위 4 이하인 경우 해당한다고 판단)
    if ("가계 곤란한" in criteria or "가정형편이 어려운" in criteria or "가정형편이 어렵고" in criteria) and student_info["소득분위"] <= 4:
        score += 1

    return score

# 사용자 프로필을 텍스트로 변환하는 함수
def student_info_to_text(info):
    return f"{info['학년']} {info['전공']} 성적 {info['성적']} 소득분위 {info['소득분위']} {' '.join(info['기타'])}"

# 학생 정보를 텍스트로 변환
student_text = student_info_to_text(student_info)

# 장학금 조건 텍스트로 변환
scholarship_texts = [scholarship["criteria"] for scholarship in scholarships]

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([student_text] + scholarship_texts)

# 코사인 유사도 계산
cosine_similarities = cosine_similarity(vectors[0:1], vectors[1:]).flatten()

# 유사도와 추가 점수를 합산하여 최종 점수 계산
final_scores = []
for i, scholarship in enumerate(scholarships):
    criteria_score = check_criteria(scholarship['criteria'], student_info)
    final_score = cosine_similarities[i]*50 + criteria_score
    final_scores.append((scholarship['name'], final_score, scholarship['criteria']))

# 최종 점수 순으로 정렬하여 추천
final_scores = sorted(final_scores, key=lambda x: x[1], reverse=True)

# 추천 장학금 목록 출력
print("추천 장학금 목록:")
for scholarship, score, criteria in final_scores[:4]:
    print(f"{scholarship} (점수: {score:.2f}) - 기준: {criteria}")
```
